```python
from tokenizers import Tokenizer
tokenizer = Tokenizer.from_pretrained("gpt2")  # Load GPT-2’s BPE tokenizer
text = "So we beat on, boats against the curret, borne back ceaselessly into the past"
encoded = tokenizer.encode(text)

print(encoded.tokens, "\n")
print(encoded.ids)
```

```python
# Embedding matrix
import numpy as np
vocab_size = tokenizer.get_vocab_size()
print(vocab_size)

d_emb = 768
embedding_matrix = np.random.uniform(low=-0.05, high=0.05, size=(vocab_size, d_emb))
print(embedding_matrix.shape)
print(embedding_matrix[0])  # Outputs: (50257, 768)

```
